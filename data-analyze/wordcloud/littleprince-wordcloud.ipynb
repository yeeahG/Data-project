```python
from wordcloud import WordCloud
```

```python
import matplotlib.pyplot as plt
```

```python
file_name = './littleprince_djvu.txt'
with open(file_name,  encoding='UTF8') as f:
    text = f.read()
```

```python
wordcloud_img = WordCloud().generate(text)
```

```python
plt.imshow(wordcloud_img, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud_img = WordCloud(background_color='white', max_font_size=300, width=1000, height=600).generate(text)
```

```python
plt.imshow(wordcloud_img, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud_img.to_file('./littleprince.png')
plt.show()
```
